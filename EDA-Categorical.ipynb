```python
import pandas as pd
dfCSV = pd.read_csv("masterfile.csv")

# Missing data
dfCSV.apply(lambda x: sum(x.isnull()),axis=0) 
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#Categorical Data: PreferredShoppingPlatform
# 1. Info stat = Frequency/Mode
# 2. Visualization tool = Bar chart
data = dfCSV['PreferredShoppingPlatform'].value_counts(ascending=True)

total_responses = len(dfCSV['PreferredShoppingPlatform'])
percentages = [(count / total_responses) * 100 for count in data]

fig, (ax1) = plt.subplots(figsize=(6, 7))

# Define different colors for each bar
colors = ['skyblue', 'lightgreen', 'lightcoral']

bars = ax1.bar(['Online shopping', 'In-store shopping', 'Both'], 
               data, color=colors)

# Annotate bars with counts and percentages
for bar, count, percentage in zip(bars, data, percentages):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width() / 2, height, 
             f'{count}\n({percentage:.2f}%)', ha='center', va='bottom')

ax1.set_title("Preferred Shopping Platform")
ax1.grid(False)
plt.tight_layout()

plt.show()
```

```python
#Categorical Data: PreferredClothingBrand, Gender
# 1. Info stat = Frequency/Mode
# 2. Visualization tool = Stacked Bar chart
bar_dfCSV = pd.crosstab(dfCSV['PreferredClothingBrand'],
                        dfCSV['Gender'])
bar_dfCSV.plot(kind='bar', stacked=True, colormap='viridis')

plt.title('Preferred Clothing Brand Based On Gender')
plt.xlabel('Clothing Brand')
plt.ylabel('Count')
plt.legend(title='Gender', bbox_to_anchor=(1.05, 1), 
           loc='upper left')

plt.show()
```

```python
#Categorical Data: PreferredClothingBrand
# 1. Info stat = Kernel Density Estimation (KDE)
# 2. Visualization tool = Violin plot
sns.violinplot(x='PreferredClothingBrand', y='Age', data=dfCSV, 
               palette=['red', 'darkorange', 'deepskyblue'])
plt.xlabel('Clothing Brand')
plt.ylabel('Age')
plt.title('Violin Plot: Age Distribution for Preferred Clothing Brand')
plt.show()
```

```python
#Categorical Data: ClothesExpenses, PreferredClothingBrand
# 1. Info stat = Median,Whiskers
# 2. Visualization tool = Box plot
sns.boxplot(x='PreferredClothingBrand', y='ClothesExpenses', 
            data=dfCSV, palette=['red', 'darkorange', 'deepskyblue'])
plt.xlabel('Preferred Clothing Brand')
plt.ylabel('Clothing Expenses')
plt.title('Box Plot: Clothing Expenses Distribution for Preferred Clothing Brand')
plt.show()
```
