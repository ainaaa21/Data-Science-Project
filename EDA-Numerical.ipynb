```python
import pandas as pd
import matplotlib.pyplot as plt


dfCSV = pd.read_csv("masterfile.csv")

# Missing data
dfCSV.apply(lambda x: sum(x.isnull()),axis=0) 
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#Numerical Data: ClothesExpenses,Age
# 1. Info stat = Median,Whiskers
# 2. Visualization tool = Box plot
sns.boxplot(x='Age', y='ClothesExpenses', data=dfCSV)

plt.xlabel('Age')
plt.ylabel('Clothing Expenses')
plt.title('Clothing Expenses based on Age')
plt.show()
```

```python
#Numerical Data: Age
# 1. Info stat = Frequency/Mode
# 2. Visualization tool = Histogram
plt.figure(figsize=(10,6))
plt.hist(dfCSV['Age'], bins=20, color='skyblue', 
         edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.show()
```

```python
#Numerical Data: ClothesExpenses,MonthlyExpenses
# 1. Info stat = Correlation
# 2. Visualization tool = Scatter plot
plt.figure(figsize=(10,6))
plt.scatter(dfCSV['MonthlyExpenses'], dfCSV['ClothesExpenses'], 
            color='skyblue')
plt.title('Comparison between Monthly Expenses and Clothes Expenses')
plt.xlabel('Monthly Expenses')
plt.ylabel('Clothes Expenses')
plt.show()
```
